```python
import pandas as pd
import datetime
from datetime import datetime
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
df = pd.read_csv('201908-citibike-tripdata.csv')
```

```python
# 2. Check the datatypes of your columns. 
df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
df['tripduration_min'] = pd.to_datetime(df['tripduration'], unit='s')
```

```python
# 4. Check the datatypes of your columns. 

df.dtypes
```


ADDITIONAL CODE

```python
# Extra: Adding age column based on year of date and birth year.
df['age'] = 2019- df['birth year']
```

```python
# Replacing gender with 0 = unknown, 1= male, 2=female
df = df.replace({'gender' : {0 : 'unknown', 1 : 'male', 2 : 'female'}})
```

```python
# Organizing columns
df = df[['tripduration','tripduration_min','starttime', 'stoptime', 'start station id', 'start station name', 'start station latitude', 'start station longitude', 'end station id', 'end station latitude', 'end station longitude', 'bikeid', 'usertype', 'birth year', 'age', 'gender']]
```

```python
# Validating DF
df.head()
```

```python
# Validating dtypes
df.dtypes
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
df.to_csv('201909-citibike_tripdata_fixed.csv', index = False)
```
